#### 2. cache

#### 4. batch

### Preprocessing techniques used to optimize dataset loading and training efficiency in machine learning workflows, especially in TensorFlow.

```python
# reloaded the dataset
ds = tfds.load('fashion_mnist', split = 'train')

# normalizes the image pixel values.
ds = ds.map(scale_images)

# caches the dataset in memory for faster access during subsequent epochs.
ds = ds.cache()

# shuffles the dataset with a buffer size of 60000. This randomizes the order of elements for training.
ds = ds.shuffle(60000)

# divides the dataset into batches of 128 elements. This is common practice in machine learning to improve training efficiency.
ds = ds.batch(128)

# prefetches 64 batches of data in the background to reduce latency during training.
ds = ds.prefetch(64)
```

## 3. Build neural network

### 3.1 Importing modeling components

```python
# Sequential API one input and one output flow
from tensorflow.keras.models import Sequential

# Bring in layers for the neural network   are fully connected layers           Regularization
from tensorflow.keras.layers import Conv2D, Dense, Flatten, Reshape, LeakyReLU, Dropout, UpSampling2D, MaxPooling2D
#                      perfrom convolutions        flattern set of vectors, dimensions, matrices

# reshape -> transform output from prev layer look like to different shape(random numbers to sth having spatial quality)
```

### 3.2 Building generator

```python
def build_generator():
    model = Sequential()

    # Takes in random values and reshapes it to 7x7x128
    # Beginnings of a generated image
    model.add(Dense(7*7*128, input_dim=128))
    model.add(LeakyReLU(0.2))
    model.add(Reshape((7,7,128)))

    # Upsampling block 1
    model.add(UpSampling2D())
    model.add(Conv2D(128, 5, padding='same'))
    model.add(LeakyReLU(0.2))

    # Upsampling block 2
    model.add(UpSampling2D())
    model.add(Conv2D(128, 5, padding='same'))
    model.add(LeakyReLU(0.2))

    # Convolutional block 1
    model.add(Conv2D(128, 4, padding='same'))
    model.add(LeakyReLU(0.2))

    # Convolutional block 2
    model.add(Conv2D(128, 4, padding='same'))
    model.add(LeakyReLU(0.2))

    # Conv layer to get to one channel
    model.add(Conv2D(1, 4, padding='same', activation='sigmoid'))

    return model
```

```python
# testing model
generator = build_generator()
```

### 3.3 Discriminator

```python
def build_discriminator():
    model = Sequential()

    # First Conv Block
    model.add(Conv2D(32, 5, input_shape=(28, 28, 1), padding='same'))  # Add padding='same'
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.4))

    # Second Conv Block
    model.add(Conv2D(64, 5, padding='same'))  # Add padding='same'
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.4))

    # Third Conv Block
    model.add(Conv2D(128, 5, padding='same'))  # Add padding='same'
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.4))

    # Fourth Conv Block
    model.add(Conv2D(256, 5, padding='same'))  # Add padding='same'
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.4))

    # Flatten then pass to dense layer
    model.add(Flatten())
    model.add(Dropout(0.4))
    model.add(Dense(1, activation='sigmoid'))

    return model
```

## 4. Construct Training loop

### 4.1 Setup losses and optimizers

```python
from tensorflow.keras.optimizers import Adam                #   optimizer for both
from tensorflow.keras.losses import BinaryCrossentropy      # loss for both
```

```python
g_opt = Adam(learning_rate=0.0001)              #   low so discriminator deosn't learn too fast
d_opt = Adam(learning_rate=0.00001)
g_loss = BinaryCrossentropy()                    # Entropy
d_loss = BinaryCrossentropy()
```

### 4.2 Build subclass model

```python
# importing base model to subclass out training step
from tensorflow.keras.models import Model
```

```python
class FashionGAN(tf.keras.Model):
    def __init__(self, generator, discriminator, latent_dim=128, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.generator = generator
        self.discriminator = discriminator
        self.latent_dim = latent_dim

    def compile(self, g_opt, d_opt, g_loss, d_loss, *args, **kwargs):
        super().compile(*args, **kwargs)
        self.g_opt = g_opt
        self.d_opt = d_opt
        self.g_loss = g_loss
        self.d_loss = d_loss

    def train_step(self, real_images):
        batch_size = tf.shape(real_images)[0]

        # Generate random noise
        random_latent_vectors = tf.random.normal(shape=(batch_size, self.latent_dim))

        # Train the discriminator
        with tf.GradientTape() as d_tape:
            # Generate fake images
            fake_images = self.generator(random_latent_vectors, training=False)

            # Get discriminator outputs
            real_output = self.discriminator(real_images, training=True)
            fake_output = self.discriminator(fake_images, training=True)

            # Add noise to labels
            real_labels = tf.ones_like(real_output) + 0.15 * tf.random.uniform(tf.shape(real_output))
            fake_labels = tf.zeros_like(fake_output) - 0.15 * tf.random.uniform(tf.shape(fake_output))

            # Calculate discriminator loss
            real_loss = self.d_loss(real_labels, real_output)
            fake_loss = self.d_loss(fake_labels, fake_output)
            total_d_loss = real_loss + fake_loss

        # Apply gradients to discriminator
        d_gradients = d_tape.gradient(total_d_loss, self.discriminator.trainable_variables)
        self.d_opt.apply_gradients(zip(d_gradients, self.discriminator.trainable_variables))

        # Train the generator
        with tf.GradientTape() as g_tape:
            # Generate fake images
            fake_images = self.generator(random_latent_vectors, training=True)

            # Get discriminator output for fake images
            fake_output = self.discriminator(fake_images, training=False)

            # Calculate generator loss
            total_g_loss = self.g_loss(tf.ones_like(fake_output), fake_output)

        # Apply gradients to generator
        g_gradients = g_tape.gradient(total_g_loss, self.generator.trainable_variables)
        self.g_opt.apply_gradients(zip(g_gradients, self.generator.trainable_variables))

        return {"d_loss": total_d_loss, "g_loss": total_g_loss}
```

```python
# Create instance of subclassed model
fashgan = FashionGAN(generator, discriminator)
```

```python
# Compile the model
fashgan.compile(g_opt, d_opt, g_loss, d_loss)
```

## 4.3. Building callback

```python
import os
from tensorflow.keras.preprocessing.image import array_to_img
from tensorflow.keras.callbacks import Callback
```

```python
import tensorflow as tf
from tensorflow.keras.callbacks import Callback
from tensorflow.keras.preprocessing.image import array_to_img
import os

class ModelMonitor(Callback):
    def __init__(self, num_img=3, latent_dim=128):
        """
        Initialize the ModelMonitor callback.

        Args:
        num_img (int): Number of images to generate and save at each epoch end.
        latent_dim (int): Dimension of the latent space for the generator.
        """
        self.num_img = num_img
        self.latent_dim = latent_dim

    def on_epoch_end(self, epoch, logs=None):
        """
        This method is called at the end of each training epoch.

        Args:
        epoch (int): The current epoch number.
        logs (dict): Dictionary containing the metrics results for this epoch.
        """
        # Generate random latent vectors
        random_latent_vectors = tf.random.uniform((self.num_img, self.latent_dim, 1))

        # Use the generator to create images from these latent vectors
        generated_images = self.model.generator(random_latent_vectors)

        # Scale the images to 0-255 range
        generated_images *= 255

        # Convert to numpy array
        generated_images = generated_images.numpy()

        # Save each generated image
        for i in range(self.num_img):
            # Convert the image array to a PIL Image object
            img = array_to_img(generated_images[i])

            # Save the image to a file
            img.save(os.path.join('/content/sample_data/images', f'generated_img_{epoch}_{i}.png'))
```

## 4.3. Train model

```python
# Recommend 2000 epochs
hist = fashgan.fit(ds, epochs=20, callbacks=[ModelMonitor()])
```

## 4.4. Review Performance
